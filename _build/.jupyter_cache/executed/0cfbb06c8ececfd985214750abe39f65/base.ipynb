```python
from sympy import symbols, sin
from sympy.plotting import plot
x = symbols('x')
funky_function = sin(x) + sin((10.0 / 3.0) * x)
p1 = plot(funky_function, show=False)
p2 = plot(funky_function.diff(x), show=False)
# p2 = plot(x, show=False)
p1.append(p2[0])
# p1
p1.show()
```
