```python
from sympy import symbols, sin, Polygon
from sympy.plotting import plot
from spb import *
import seaborn as sns
sns.set_theme()

x = symbols('x')
funky_function = sin(x) + sin((10.0 / 3.0) * x)
p1 = plot(funky_function, show=False)
p2 = plot(funky_function.diff(x), show=False)
# p2 = plot(x, show=False)
p1.append(p2[0])
# p1
p3 = graphics(
    geometry(Polygon((4, 0), 4, n=8)),
    aspect="equal", grid=False, show=False
)
p1.append(p3[0])

# p1.show()
```
